```python
import random
random.seed(401)
a = []
for i in range(1,100):
    a.append(random.randint(1,100000))
print(a)
```

```python
def linear(li,key):
    flag = 0
    for i in li:
        if i==key:
            return 1
    if flag==0:
        return 0
```

```python
%%time
#TIME TAKEN FOR NUMBER IS LESS AS LIST IS ONLY 100 ELEMENTS
print(linear(a, 5))
```

```python
import random
random.seed(401)
a = []
for i in range(1,100000):
    a.append(random.randint(1,100000))
print(a)
```

```python
%%time
#TIME TAKEN FOR NUMBER IS MUCH LARGER AS LIST IS LARGE WITH 100000 ELEMENTS
print(linear(a, 99999))
```

```python
#CONCLUSION
#LINEAR SEARCH SPEED DEPENDS ON THE SEARCH SPACE AS IT SCANS EACH AND EVERY ELEMENT LINEARLY TILL IT FINDS THE
#REQUIRED KEY.
```
